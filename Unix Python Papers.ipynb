### Some Cool Stuff Python Can Do

```python
# Astronomy package
from astropy.io import fits 
```

```python
file_dir = "data/"
img_name = "pdi_pi_collapsed.fits"
img = fits.getdata(file_dir + img_name, ext = 1)
```

```python
#create a figure
fig = plt.figure(figsize = (8,8))
image = plt.imshow(img, origin = "lower")
cbar = fig.colorbar(image, shrink = 0.82)
cbar.set_label("Surface Brightness (arb. units)", rotation = 90, fontsize = 12)
plt.xlim([50,160])
plt.ylim([50,160])
plt.xlabel(r'$\Delta$ RA [px]', fontsize = 12)
plt.ylabel(r'$\Delta$ Dec [px]', fontsize = 12);
```

```python
import pandas as pd 
```

```python
stellar = pd.read_csv(file_dir + "closest20kstars.csv")
```

```python
stellar
```

```python
# create a fig from stellar
fig = plt.figure(figsize = [10,10]) 
plt.scatter(stellar["ra"], stellar["dec"], alpha = 0.8, s = 12, lw = 0.5, ec = "k", color = "darkorchid")
plt.xlim([0,362])
plt.ylim([-75,75])
plt.xlabel("RA [$\degree$]", fontsize = 14)
plt.ylabel("Dec [$\degree$]", fontsize = 14)
plt.title("Closest Stars 20.000 stars from Gaia Archive (DR2)", fontsize = 18)
```

```python
#save figure
fig.savefig("stars.png", dpi = 300)
```

```python
# Fibonacci
```

```python
def calculate_fibonacci(num_terms):
    sequence = [0,1]
    
    for i in range(2, num_terms):
        next_term = sequence[-1] + sequence[-2]
        
        sequence.append(next_term)
        
    return sequence
```

```python
num_terms = int(input("enter the number of terms in the fibonacci sequence: "))
fib_sequence = calculate_fibonacci(num_terms)
```
